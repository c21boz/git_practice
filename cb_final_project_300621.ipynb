The following information is based on the Open University Learning Analytics Dataset (OULAD): https://www.kaggle.com/anlgrbz/student-demographics-online-education-dataoulad Accessed: 19.06.21

Objective:
To investigate the Open University studentInfo data to identify any trends in final outcomes based on gender, disability or age categories.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# import studentInfo.csv
dfstudent = pd.read_csv ('~/Datascience/OUdata/studentInfo.csv')
```

```python
#create new dataframe of gender and final_result
df1 = dfstudent.iloc[:,[3,11]]
print (df1)
```

```python
df1.head()
```

```python
#plot simple bar chart for gender variable
df1['gender'].value_counts().plot(kind='bar', title = 'Number of students by gender')
#add x label and y label 
plt.xlabel('Gender')
plt.ylabel('Number of students')
```

```python
#plot simple bar chart for final results variable and change bar color
df1['final_result'].value_counts().plot(kind='bar', color = 'hotpink')
#Name xlabel, ylabel and title
plt.title('Number of students per final result outcome')
plt.xlabel('Final result outcome')
plt.ylabel('Number of students')
```

```python
#plot simple bar chart for final results variable and change bar color
dfstudent['age_band'].value_counts().plot(kind='bar', color = 'brown')
#Name xlabel, ylabel and title
plt.title('Number of students per age band')
plt.xlabel('Age Band')
plt.ylabel('Number of students')
```

```python
#plot simple bar chart for final results variable and change bar color
df1['final_result'].value_counts().plot(kind='bar', color = ['hotpink', 'blue', 'green', 'purple'])
#Name xlabel, ylabel and title
plt.title('Number of students per final result outcome')
plt.xlabel('Final result outcome')
plt.ylabel('Number of students')
#rotate the x labels 
plt.xticks(rotation=30, horizontalalignment="center")
```

```python
#compare outcomes per category by gender with catplot
sns.set_theme(style="whitegrid")
sns.catplot(x="final_result", kind="count", hue='gender', palette="pastel", edgecolor='0.6', data=dfstudent)
```

```python
#compare outcomes per category by disability status with catplot
sns.set_theme(style="whitegrid")
sns.catplot(x="final_result", kind="count", hue='disability', palette="ch:.50", edgecolor='0.6', data=dfstudent)
```

```python
#compare numbers of students per previous academic attainment category
sns.catplot(x="highest_education", kind="count", hue='gender', palette="ch:.25", edgecolor='0.6', data=dfstudent)
plt.xticks(rotation=45)
```

```python
#create a bar graph to show outcomes by age band and gender
sns.catplot(x="age_band", col="final_result", col_wrap=4,data=dfstudent, kind="count", 
            hue='gender', palette="Set2", edgecolor='0.6', height=5, aspect=0.6)
           
```
